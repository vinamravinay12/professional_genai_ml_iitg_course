```python
def null_check(data):
    null_counts = data.isnull().sum()
null_check(marketingData)

def clean_categorical_data(data):
    data.columns = (
    data.columns
      .str.strip()          # remove leading/trailing spaces
      .str.replace(" ", "")# replace inner spaces with _
      .str.replace("-", "_")# replace hyphens if any
               # make all lowercase (optional)
)
    data["Education"] = data["Education"].str.strip().str.title()
    data["Marital_Status"] = data["Marital_Status"].str.strip().str.title()
    print("Cleaned 'Education' and 'Marital_Status' columns.")
    print('education data : ', data['Education'].value_counts())
    print('marital status data : ', data['Marital_Status'].value_counts())


clean_categorical_data(marketingData)
```

```python
def convert_income_to_float(data): 
    data["Income"] = data["Income"].astype("str").str.replace(r"[\$,]", "", regex=True).astype(float)
    print("dtype of income: ", data["Income"].dtype)
    

convert_income_to_float(marketingData)

def get_missing_income_values_count(data) :
    return data["Income"].isna().sum()

missing_income_values = get_missing_income_values_count(marketingData)
print("missing income : ", missing_income_values)

##Compute average income for each group based on Marital status and Education

def get_average_income_on_group_by_marital_status_and_education(data):
    group_means = (
    data.groupby(["Education", "Marital_Status"])["Income"]
      .mean()
    )
    return group_means
 
average_income_on_group = get_average_income_on_group_by_marital_status_and_education(marketingData)
##print(average_income_on_group)

##fill missing income data based on
def fill_missing_income_data_based_on_average(data): 
    data["Income"] = data["Income"].fillna(
    data.groupby(["Education", "Marital_Status"])["Income"]
      .transform("mean")
)

    

fill_missing_income_data_based_on_average(marketingData)
```
