```python
customers = ["Ravi", "Vinamra", "Vinay", "Ritika","Shubham","Aman"]

orders = [
    ("Vinamra","Laptop", 900.00, "Electronics"),
     ("Ravi", "Laptop", 600.00, "Electronics"),
    ("Vinay", "Keyboard", 40.00, "Electronics"),
    ("Shubham", "Shirt", 20.00, "Clothing"),
    ("Ritika", "Jeans", 30.00, "Clothing"),
    ("Ravi", "Microwave", 100.00, "Appliances"),
    ("Ritika", "Air Fryer", 80.00, "Appliances"),
    ("Vinamra", "Airpods", 150.00, "Electronics"),
     ("Vinay", "Shoes", 50.00, "Clothing"),
    ("Shubham", "Vaccum Cleaner", 60.00, "Appliances"),
    ("Aman","Shirt", 30.00, "Clothing"),
    ("Vinamra","Blazer",70.00,"Clothing")
]

product_to_category = {
    "Laptop" : "Electronics",
    "Keyboard": "Electronics",
    "Airpods": "Electronics",
    "Shirt": "Clothing",
    "Jeans": "Clothing",
    "Shoes" : "Clothing",
    "Vacuum Cleaner": "Appliances",
    "Microwave": "Appliances",
    "Vaccum Cleaner": "Appliances",
}

HIGH_VALUE_MIN = 100.00
MODERATE_MIN = 50.00
customer_to_products = {}
for customer, product, price, category in orders: 
    customer_to_products.setdefault(customer, []).append(product)

unique_categories = set(product_to_category.values())
customer_spending = {}
for name, product, amount, category in orders :
    customer_spending[name] = customer_spending.get(name,0.0) + amount 

#classifying high value customers 
customer_classification = {}
for name, total in customer_spending.items() :
      if total > HIGH_VALUE_MIN : 
        customer_classification[name] = "High-Value"

      elif total >= MODERATE_MIN : 
         customer_classification[name] = "Moderate"

      else :
         customer_classification[name] = "Low-Value"


#Revenue per category for products 
category_revenue = {}
for name, product, price, category in orders : 
    category_revenue[category] = category_revenue.get(category,0.0) + price


#unique products across categories 
unique_products = { product for _,product,_,_ in orders}

#all electronics buyers 
electronics_buyers = [name for name, _,_,category in orders if category == "Electronics"]

#top 3 spenders 
top_spenders = sorted(customer_spending.items(),
                      key = lambda key : key[1],
                      reverse=True)

high_value_customers = sorted(
    [name for name, title in customer_classification.items() if title  == "High-Value"]
)

product_counts = {}
for _, product, _, _ in orders:
    product_counts[product] = product_counts.get(product, 0) + 1


# Multiple category buyers
customer_to_categories = {}
for name,_,_,category in orders : 
    customer_to_categories.setdefault(name,[]).append(category)


multiple_category_buyers = {name for name,categories in customer_to_categories.items() if len(categories) > 1 }

# customers who brought both electronics and clothing
customer_with_electronics_and_clothing = { name for name, category in customer_to_categories.items()
     if {"Electronics", "Clothing"}.issubset(category)
    }


#print insight title
def print_title(title: str):
    print("\n" + title)
    print("-" * len(title))

def print_table(rows, headers=None, col_sep="  "):
    rows = list(rows)
    if headers:
        rows = [headers] + rows
    # Compute column widths
    widths = []
    for col_idx in range(len(rows[0])):
        col_values = [str(r[col_idx]) for r in rows]
        widths.append(max(len(v) for v in col_values))
    # Print
    for i, r in enumerate(rows):
        line = col_sep.join(str(val).ljust(widths[j]) for j, val in enumerate(r))
        print(line)
        if headers and i == 0:
            print(col_sep.join("-" * w for w in widths))


def main():
    # Section: Inputs overview
    print_title("Available Product Categories")
    print(", ".join(sorted(unique_categories)))

    print_title("Unique Products")
    print(", ".join(sorted(unique_products)))

    # Section: Customer summaries
    print_title("Customer Spending & Classification")
    rows = []
    for name in sorted(customer_spending.keys()):
        rows.append((name, f"${customer_spending[name]:.2f}", customer_classification[name]))
    print_table(rows, headers=("Customer", "Total Spending", "Classification"))

    # Section: Category revenue
    print_title("Revenue by Category")
    rows = []
    for category in sorted(category_revenue.keys()):
        rows.append((category, f"${category_revenue[category]:.2f}"))
    print_table(rows, headers=("Category", "Revenue"))

    # Section: Electronics buyers (could contain duplicates if they ordered multiple electronics)
    print_title("Customers Who Purchased Electronics (may include repeats)")
    if electronics_buyers:
        print(", ".join(electronics_buyers))
    else:
        print("None")

    # Section: Top spenders
    print_title("Top 3 Highest-Spending Customers")
    rows = [(name, f"${total:.2f}") for name, total in 
            top_spenders]
    print_table(rows, headers=("Customer", "Total Spending"))

    # Section: Set-based insights
    print_title("Customers Purchasing from Multiple Categories")
    print(", ".join(sorted(multiple_category_buyers)) if multiple_category_buyers else "None")

    print_title("Customers Who Bought both Electronics and Clothing")
    print(", ".join(sorted(customer_with_electronics_and_clothing)) if customer_with_electronics_and_clothing else "None")

    print_title("High-Value Customers (> $100 total)")
    print(", ".join(high_value_customers) if high_value_customers else "None")

    K = 5
    top_k_products = sorted(product_counts.items(), key=lambda kv: kv[1], reverse=True)[:K]

    print_title(f"Most Frequently Purchased Products (Top {K})")
    print_table(top_k_products, headers=("Product", "Count"))
if __name__ == "__main__":
    main()
```
